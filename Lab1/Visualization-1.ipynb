# 3. Data Visualization

### 3.1 Data Exploration



```python
concentration_votes_2016 = votes.groupby(by="state_abbr").sum()[["votes_dem_2016","votes_gop_2016","total_votes_2016","population2014","AGE295214"]]
concentration_votes_2016.sort_values(by=["total_votes_2016"])
```

#### 3.1.1 Total Vote Frequency by State (2016)

```python
import plotly.graph_objects as go
import pandas as pd

percent_voted = (concentration_votes_2016['total_votes_2016']/ \
    (concentration_votes_2016['population2014']-concentration_votes_2016['AGE295214'])) * 100

text ='Democratic Votes: ' + concentration_votes_2016['votes_dem_2016'].astype(str) + '<br>' + \
    'Republican Votes: ' + concentration_votes_2016['votes_gop_2016'].astype(str) + '<br>' + \
    'Total Votes: ' + concentration_votes_2016['total_votes_2016'].astype(str) + '<br>' + \
    '2014 Population: ' + concentration_votes_2016['population2014'].astype(str) + '<br>' + \
    'Pecent Voted: ' + (percent_voted.round(3)).astype(str) + '%<br>' + \
    'Minimum count: ' + str(concentration_votes_2016['total_votes_2016'].min()) + '<br>' + \
    'Maximum count: ' + str(concentration_votes_2016['total_votes_2016'].max())

fig1 = go.Figure(
    data=go.Choropleth(
        locations=concentration_votes_2016.index,
        z = concentration_votes_2016['total_votes_2016'].astype(int),
        locationmode = 'USA-states',
        colorscale = 'greys',
        colorbar_title = "# of voters (million)",
        text=text
    )
)

fig1.update_layout(
    title_text = 'Concentration of Total Votes By State',
    geo_scope='usa',
)

fig1.show()
```

The concentration of total votes by state visualization gives us a good understanding of where the majority of votes are coming from accross the U.S. The highest vote count was 11.73 million people in CA and the lowest was 248,742 people in WY followed by DC and then VT.

#### 3.1.4 Clinton vs. Trump & Obama vs. Romney


```python
import matplotlib.pyplot as plt

new = votes.groupby(by="state_abbr").sum()[["votes_dem_2016","votes_gop_2016","total_votes_2016","votes_dem_2012","votes_gop_2012","total_votes_2012"]]
new_sum = new.sum()
result_percent = pd.DataFrame(index={"% of vote"})

result_percent["clinton"] = (new_sum["votes_dem_2016"] / new_sum["total_votes_2016"]).astype(float)
result_percent["trump"] = (new_sum["votes_gop_2016"] / new_sum["total_votes_2016"]).astype(float)
result_percent["obama"] = (new_sum["votes_dem_2012"] / new_sum["total_votes_2012"]).astype(float)
result_percent["romney"] = (new_sum["votes_gop_2012"] / new_sum["total_votes_2012"]).astype(float)

print(result_percent)

result_percent.T.plot(kind="bar",)
plt.ylim(top=.6)
plt.title('Percent Votes By Candidate')
plt.ylabel('Pecent of Vote')
plt.xlabel('Candidates')
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.show()
```

The percentage of votes won by candidate bar graph includes the 2012 and 2016 election data of Clinton vs. Trump and Obama vs. Romney. We see above that Obama won the popular vote by ~3% in the 2012 election while in 2016, Clinton actually won the popular vote by ~1%. Even though Clinton won the popular vote Trump still won the office due to the electoral college. This is interesting because it may have something to say about how the electoral college functions and also that one can win the presidency by factors other than the popular vote.

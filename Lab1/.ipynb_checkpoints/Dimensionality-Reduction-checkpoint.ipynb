```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

from sklearn.manifold import TSNE

pd.set_option('display.max_columns',500)
pd.set_option('display.max_rows',500)
```

```python
#read in data
votes = pd.read_csv("./2012-and-2016-presidential-elections/votes.csv")
```

### Data Pre-Processing

```python

attributes = ['votes_dem_2016',
              'votes_gop_2016',
              'total_votes_2016',
              'Clinton',
              'Trump',
              'area_name',
              'state_abbr',
              'votes_dem_2012',
              'votes_gop_2012',
              'total_votes_2012',
              'Obama','Romney',
              'population2014',
              'SEX255214',
              'White',
              'Black',
              'RHI325214',
              'RHI425214',
              'RHI525214',
              'RHI625214',
              'Hispanic',
              'RHI825214',
              'Edu_highschool',
              'Edu_batchelors',
              'INC110213',
              'Poverty',
              'Trump_Prediction',
              'Clinton_Prediction']

votes = votes[attributes]

#add target attribute
votes['Party_Prediction'] = np.where(votes['Trump']> 0.5, 1,0)
votes
```

```python
votes.Party_Prediction.hist()
```

### One-hot-encoding

```python
votes_encoded = pd.get_dummies(votes)
votes_encoded.shape
```

### t-SNE Dimensionality Reduction with 2 dimensions

t-Distributed Stochastic Neighbor Embedding (t-SNE) is a dimensionality reduction method that is used to reduce the dimensions of data from a high dimensional space to a low dimensional space. t-SNE models each high dimensional attribute to a lower dimensional point where similar points will be mapped by nearby points and disimilar points are mapped by distant points. The method will map a probability distribution over the high dimensional data where observations that are similar will have a higher probability of being chosen compared to disimilar objects. [1]

```python
tsne = TSNE(n_components=2,verbose=1,perplexity=10, n_iter=300)
tsne_results = tsne.fit_transform(votes_encoded)
```

```python
votes_encoded['1st Dimension'] = tsne_results[:,0]
votes_encoded['2nd Dimension'] = tsne_results[:,1]
plt.figure(figsize=(16,10))
sns.scatterplot(
    x="tsne-2d-one", y="tsne-2d-two",
    hue="Party_Prediction",
    data=votes_encoded,
    legend="full",
    alpha=0.3
)
```

```python
tsne = TSNE(n_components=3,verbose=1,perplexity=10, n_iter=300)
tsne_results = tsne.fit_transform(votes_encoded)
```

```python
votes_encoded['1st Dimension'] = tsne_results[:,0]
votes_encoded['2nd Dimension'] = tsne_results[:,1]
votes_encoded['3rd Dimension'] = tsne_results[:,2]

plt.figure(figsize=(16,10))
sns.scatterplot(
    x="1st Dimension", y="2nd Dimension",
    hue="Party_Prediction",
    data=votes_encoded,
    legend="full",
    alpha=0.3
)
```

```python
ax = plt.figure(figsize=(16,10)).gca(projection='3d')
colors = {1:'red',0:'blue'}
ax.scatter(
    xs=votes_encoded["1st Dimension"], 
    ys=votes_encoded["2nd Dimension"], 
    zs=votes_encoded["3rd Dimension"], 
    c=votes_encoded["Party_Prediction"].apply(lambda x: colors[x]), 
    cmap='tab10'
)
ax.set_xlabel('1st Dimension')
ax.set_ylabel('2nd Dimension')
ax.set_zlabel('3rd Dimension')
ax.legend(['Republican','Democrat'])
plt.show()
```
